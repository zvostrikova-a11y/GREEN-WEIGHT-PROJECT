Кафе "Авокадо" (салаты)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import re

print("AVOCADO CAFE (ЦЕНА, ВЕС, ЦЕНА ЗА 100г)")
print("="*60)

options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.binary_location = '/usr/bin/chromium-browser'

driver = webdriver.Chrome(options=options)
driver.get("https://avocadocafe.ru/takeaway#!/tab/683817402-1")
time.sleep(8)

text = driver.find_element(By.TAG_NAME, "body").text
blocks = re.split(r'Заказать', text)

dishes = []

for block in blocks:
    if '₽' not in block:
        continue

    # Цена
    price_match = re.search(r'(\d+)\s*₽', block)
    if not price_match:
        continue
    price = int(price_match.group(1))

    # Вес
    weight_match = re.search(r'(\d+)\s*гр', block)
    if not weight_match:
        continue
    weight = int(weight_match.group(1))

    # Цена за 100г
    price_per_100g = round((price / weight) * 100, 1)

    # Название
    lines = block.strip().split('\n')
    name = None
    for line in lines:
        line = line.strip()
        if (len(line) > 5 and re.match(r'[А-ЯЁ]', line) and 'Артикул' not in line and not re.match(r'\d', line)):
            name = line
            break

    if name and price and weight:
        dishes.append({
            'name': name,
            'price': price,
            'weight_g': weight,
            'price_per_100g': price_per_100g
        })

driver.quit()


print("МЕНЮ AVOCADO CAFE")
for d in dishes:
    print(f"{d['name']}")
    print(f"   {d['price']} ₽ | {d['weight_g']} г | {d['price_per_100g']} ₽ за 100г\n")

# статистика
if dishes:
    prices_per_100g = [d['price_per_100g'] for d in dishes]
    avg = sum(prices_per_100g) / len(prices_per_100g)
    min_price = min(prices_per_100g)
    max_price = max(prices_per_100g)

    print("СТАТИСТИКА")
    print(f"Средняя цена за 100г: {avg:.1f} ₽")
    print(f"Минимум за 100г: {min_price:.1f} ₽")
    print(f"Максимум за 100г: {max_price:.1f} ₽")

    # сравнение с нашим проектом
    print("\n" + "="*60)
    print("СРАВНЕНИЕ С GREEN-WEIGHT-PROJECT")
    our_price = 120
    print(f"Наша цена за 100г: {our_price} ₽")
    print(f"Средняя цена Avocado: {avg:.1f} ₽")
    print(f"Мы дешевле на {avg - our_price:.1f} ₽ за 100г")
```

Статический парсинг (request)


```python
import requests
from bs4 import BeautifulSoup
import re

class GrabliSaladParser:

    def __init__(self, url):
        self.url = url
        self.session = requests.Session()
        self.session.headers.update({"User-Agent": "Mozilla/5.0"})
        self.salads = []

    def _fetch_page(self):
        response = self.session.get(self.url, timeout=15)
        response.raise_for_status()
        return response.text

    def _is_salad(self, text):
        text_lower = text.lower()
        exclude = ['блин', 'творог', 'брускетта', 'хлеб', 'соус', 'закуска', 'горячее']
        for word in exclude:
            if word in text_lower:
                return False
        return 'салат' in text_lower

    def _extract_name(self, text):
        lines = text.split('\n')
        name = lines[0].strip()
        if len(name) < 3 or len(name) > 80:
            match = re.search(r'([А-Я][а-яё\s]{5,50}салат[А-Яа-яё\s]*)', text, re.IGNORECASE)
            if match:
                name = match.group(1).strip()
        return name[:50]

    def parse(self):
        html = self._fetch_page()
        soup = BeautifulSoup(html, 'html.parser')
        items = soup.find_all('div', class_=re.compile('item|card|dish'))

        for item in items:
            text = item.get_text()
            if not self._is_salad(text):
                continue
            price_match = re.search(r'(\d+)\s*₽', text)
            if not price_match:
                continue
            name = self._extract_name(text)
            price = int(price_match.group(1))
            self.salads.append({'name': name, 'price': price})

        unique = []
        seen = set()
        for s in self.salads:
            if s['name'] not in seen:
                seen.add(s['name'])
                unique.append(s)
        self.salads = unique
        return self.salads

    def print_menu(self):
        if not self.salads:
            print("Салаты не найдены")
            return
        print(f"\nСАЛАТЫ GRABLI ({len(self.salads)} шт.)")
        print("-"*40)
        for s in self.salads:
            print(f"{s['name']} — {s['price']} ₽")

    def compare_with_project(self, our_price_per_100g=120, assumed_weight=150):
        if not self.salads:
            print("Нет данных для сравнения")
            return
        prices = [s['price'] for s in self.salads]
        avg_price = sum(prices) / len(prices)
        our_avg = our_price_per_100g * (assumed_weight / 100)

        print("\n" + "-"*40)
        print("СРАВНЕНИЕ С GREEN-WEIGHT-PROJECT")
        print("-"*40)
        print(f"Средняя цена Grabli: {avg_price:.0f} ₽ (при весе ~{assumed_weight}г)")
        print(f"GREEN-WEIGHT (за {assumed_weight}г): {our_avg:.0f} ₽")
        print(f"Разница: {avg_price - our_avg:.0f} ₽")

        if avg_price > our_avg:
            print(f"Мы дешевле на {avg_price - our_avg:.0f} ₽")
        else:
            print(f"Конкурент дешевле на {our_avg - avg_price:.0f} ₽")


if __name__ == "__main__":
    parser = GrabliSaladParser("https://yandex.com/navi/org/grably/12287015836/menu/")
    parser.parse()
    parser.print_menu()
    parser.compare_with_project()
```

- Цена — основная боль. Клиенты считают, что качество/количество не соответствует цене (оверпрайс)
- Вкус — жалобы на пресную, безвкусную, сухую еду.
- Общее разочарование — «типичное блогерское заведение», «не советую»

**Как GREEN-WEIGHT решает эти проблемы:**
1) Проблема Salatnica: высокие цены, оверпрайс

   Наше решение: 120 ₽/100г — прозрачная цена, без наценки за блогерский хайп

2) Проблема Salatnica: пресная / безвкусная еда
   
   Наше решение: контроль качества, свежие ингредиенты (см. раздел Тани "Контрагенты", отзывчивость к клиентам)

3) Проблема Salatnica: завышенные ожидания vs реальность (несоответствие)

   Наше решение: оплата за вес — клиент платит только за то, что реально взял

```python
import requests
import re

print("ПАРСИНГ CHECK.TOCHKA.COM (SALATNICA) - аналог rusprofile, который не спарсился")
print("="*60)

url = "https://check.tochka.com/company/1247700618945/"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
html = response.text


inn = re.search(r'ИНН\s*(\d+)', html)
print(f"ИНН: {inn.group(1) if inn else 'не найден'}")


ogrn = re.search(r'ОГРН\s*(\d+)', html)
print(f"ОГРН: {ogrn.group(1) if ogrn else 'не найден'}")

# она вообще не закрылась еще там?..
status = re.search(r'Действующая', html)
print(f"Статус: {'Действующая' if status else 'не найден'}")

# выручка 2024
revenue = re.search(r'Выручка.*?(\d[\d\s]+)\s*₽', html)
if revenue:
    print(f"Выручка (2024): {revenue.group(1).strip()} ₽")

# прибыль 2024
profit = re.search(r'Чистая прибыль.*?(\d[\d\s]+)\s*₽', html)
if profit:
    print(f"Прибыль (2024): {profit.group(1).strip()} ₽")
```

```python


revenue_val = int(re.sub(r'[^\d]', '', revenue.group(1)))
profit_val = int(re.sub(r'[^\d]', '', profit.group(1)))

plt.figure(figsize=(8, 8))
labels = ['Прибыль', 'Расходы и налоги']
sizes = [profit_val, revenue_val - profit_val]
colors = ['#4CAF50', '#FF6B6B']
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title(f'Соотношение прибыли к выручке Salatnica (2024)\nВыручка: {revenue_val:,} ₽'.replace(',', ' '), fontsize=14)
plt.tight_layout()
plt.show()

print(f"Выручка: {revenue_val:,} ₽".replace(',', ' '))
print(f"Прибыль: {profit_val:,} ₽".replace(',', ' '))
print(f"Расходы и налоги: {revenue_val - profit_val:,} ₽".replace(',', ' '))
print(f"Рентабельность: {profit_val/revenue_val*100:.1f}%")
```

Что это значит для нас?

1) Рынок прибыльный

Salatnica вышла на прибыль 7,8% уже в первый неполный год
(-> ниша востребована)

2) Высокая доля расходов —  слабое место Salatnica

У Salatnica очень высокая доля расходов относительно выручки.
Это может быть связано с: дорогой арендой в центре Москвы, высокими зарплатами персонала

3) Наше преимущество — более низкая доля расходов

За счёт формата «шведский стол + оплата за вес 120 ₽/100г » мы можем: снизить расходы на персонал (меньше поваров), уменьшить пищевые отходы (клиент платит только за что взял), оптимизировать аренду (локации рядом с фитнес-клубами дешевле центра)



ALL IN ALL

1) Ценовая стратегия

120 ₽/100г — дешевле всех конкурентов + оплата за вес — клиент платит только за то, что реально взял + нет наценки за «блогерскость» и недоеденные остатки

2) География

Локации рядом с фитнес-клубами и офисами (не центр Москвы)

3) Продуктовая стратегия

Шведский стол из 20+ ингредиентов + КБЖУ на каждом ингредиенте + контроль качества — свежие продукты (об этом дальше в разделе Тани "Контрагенты")

4) Добавляем опцию оплаты наличкой (в Salatnica жаловались, что только картой принимают)
